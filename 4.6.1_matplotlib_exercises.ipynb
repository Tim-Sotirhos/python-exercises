```python
import matplotlib.pyplot as plt
```

```python
import math
```

```python
#1) Plot the following equation: y = x^2 - x + 2

x = range(-10,12)

y = [i**2 - i + 2 for i in x]

plt.plot(x,y)
plt.annotate('(0,0)', xy = (0,0), xytext = (0,10))
plt.show()
```

```python
#2a) Create and label a chart for the following equation: y = √x 

x = range(0,100)

y = [math.sqrt(i) for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = \sqrt{x}$')

plt.show()
```

```python
#2b) Create and label a chart for the following equation: y = x^3

x = range(-5, 10)

y = [i**3 for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = x^3$')

plt.show()
```

```python
#2c) Create and label a chart for the following equation: y = tan(x)

x = range(-5, 5)

y = [math.tan(i) for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = \\tan(x)$')

plt.show()
```

```python
#2d) Create and label a chart for the following equation: y = 2^x

x = range(-10, 10)

y = [2**i for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = 2^x$')

plt.show()
```

```python
#3) Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize = (14, 7))

# plot the first subplot
plt.subplot(2,2,1)

x = range(0,100)

y = [math.sqrt(i) for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = \sqrt{x}$')

# the second subplot
plt.subplot(2,2,2)

x = range(-5, 10)

y = [i**3 for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = x^3$')

# the third subplot
plt.subplot(2,2,3)

x = range(-5, 5)

y = [math.tan(i) for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = \\tan(x)$')

# the fourth subplot
plt.subplot(2,2,4)

x = range(-10, 10)

y = [2**i for i in x]

plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y = 2^x$')

plt.suptitle('Subplots of All 4 Graphs')

plt.show()
```

```python
#4) Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points and include a legend.

plt.figure(figsize = (9, 7))

x1 = range(0,100)
y1 = [math.sqrt(i) for i in x1]
plt.plot(x1, y1, c = 'lawngreen', label = '$y = \sqrt{x}$')

x2 = x = range(-10, 100)
y2 = [i**3 for i in x2]
plt.plot(x2, y2, c = 'cornflowerblue', label = '$y = x^3$')

x3 = range(-5, 5)
y3 = [math.tan(i) for i in x3]
plt.plot(x3, y3, c='deeppink', label = '$y = \\tan(x)$')

x4 = x = range(-10, 10)
y4 = [2**i for i in x4]
plt.plot(x4, y4, c = 'darkorange', label = '$y = 2^x$')

plt.legend(loc = 'upper right')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.title('Graph of All 4 equations')
plt.ylim(-5,5)

plt.show()
```
